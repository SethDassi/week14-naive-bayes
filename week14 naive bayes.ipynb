```python
import numpy as np
import pandas as pd
import seaborn as nb
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
df=pd.read_csv('golfdataset.csv')
df
```

```python
# encoding the dataset
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df['Outlook']=le.fit_transform(df.Outlook.values)
df['Temp']=le.fit_transform(df.Temp.values)
df['Humidity']=le.fit_transform(df.Humidity.values)
df['Windy']=le.fit_transform(df.Windy.values)
df['PlayGolf']=le.fit_transform(df.PlayGolf.values)
df
```

```python
#spliting the dataset
from sklearn.model_selection import train_test_split
Labels = df['PlayGolf']
Features=df.drop('PlayGolf',axis=1)
x_train, x_test, y_train, y_test= train_test_split(Features, Labels, test_size=0.2, random_state=0)
Features
```
